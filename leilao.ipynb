```python
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score ,f1_score ,make_scorer,confusion_matrix, ConfusionMatrixDisplay,roc_curve, roc_auc_score
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tabela_treino.fillna(0,inplace=True)
tabela_test.fillna(0,inplace=True)
```

```python
X_train = tabela_treino.drop(columns='AVG(t.resultado)')
y_train = tabela_treino['AVG(t.resultado)']

X_test = tabela_test.drop(columns='AVG(t.resultado)')
y_test = tabela_test['AVG(t.resultado)']
```

```python
# Definir a métrica F1 para a validação cruzada
f1_scorer = make_scorer(f1_score, average='weighted')
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy="median")
imputer.fit(X_train)
X_train = imputer.transform(X_train)
X_test = imputer.transform(X_test)
```

```python
modelo = RandomForestClassifier(n_estimators=100, random_state=8290,class_weight={0:1,1:4})
cv_scores = cross_val_score(modelo, X_train, y_train, cv=5, scoring=f1_scorer)


print("Scores da validação cruzada:",  cv_scores)
print("Média da validação cruzada:", cv_scores.mean())
```

```python
from sklearn.metrics import brier_score_loss
```

```python
train_brier_score = brier_score_loss(y_train, modelo.predict_proba(X_train)[:, 1])

# Calculate Brier score on test data
test_brier_score = brier_score_loss(y_test, modelo.predict_proba(X_test)[:, 1])

# Print the results
print(f"Train Brier Score: {train_brier_score:.3f}")
print(f"Test Brier Score: {test_brier_score:.5f}")


train_brier_score2 = brier_score_loss(y_train, modelo2.predict_proba(X_train)[:, 1])

# Calculate Brier score on test data
test_brier_score2 = brier_score_loss(y_test, modelo2.predict_proba(X_test)[:, 1])

# Print the results
print(f"Train Brier Score: {train_brier_score2:.3f}")
print(f"Test Brier Score: {test_brier_score2:.3f}")


train_brier_score3 = brier_score_loss(y_train, modelo3.predict_proba(X_train_selected)[:, 1])

# Calculate Brier score on test data
test_brier_score3 = brier_score_loss(y_test, modelo3.predict_proba(X_test_selected)[:, 1])

# Print the results
print(f"Train Brier Score: {train_brier_score3:.3f}")
print(f"Test Brier Score: {test_brier_score3:.5f}")
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python

params = {
    "n_estimators": [100, 200, 300, 400, 500],
    "max_depth": [None, 5, 10, 15, 20],
    "min_samples_split": [2, 5, 10, 15, 20],
    "min_samples_leaf": [1, 2, 4, 6, 8],
    "class_weight": ["balanced", "balanced_subsample"],
}

# Define the objective function
def objective(params):
    model = RandomForestClassifier(**params)
    model.fit(X_train, y_train)
    y_proba = model.predict_proba(X_train)[:, 1]
    score = brier_score_loss(y_train, y_proba)
    return score

# Initialize the random search
random_search = RandomizedSearchCV(
    estimator=RandomForestClassifier(),
    param_distributions=params,
    n_iter=100,
    scoring="neg_brier_score",  # minimize Brier score
    cv=5,
    n_jobs=-1,
)


```

```python
# Fit the model
random_search.fit(X_train, y_train)

# Print the best hyperparameters
print("Best hyperparameters:", random_search.best_params_)

# Get the best model
best_model = random_search.best_estimator_

# Evaluate the best model on the test set
y_proba_test = best_model.predict_proba(X_test)[:, 1]
test_score = brier_score_loss(y_test, y_proba_test)
print("Brier score on the test set:", test_score)
```
